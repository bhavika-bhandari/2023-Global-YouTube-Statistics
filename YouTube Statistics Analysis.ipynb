```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Global_YouTube_Statistics.csv", encoding="latin-1")  # Replace with the actual path to your CSV file
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Get basic statistics about the dataset
print(data.describe())
```

```python
# Calculate the average number of subscribers
avg_subscribers = data["subscribers"].mean()
print("Average Subscribers:", avg_subscribers)
```

```python
# Calculate the total video views
total_video_views = data["video views"].sum()
print("Total Video Views:", total_video_views)
```

```python
# Find the top 5 channels with the most subscribers
top_subscribers = data.nlargest(5, "subscribers")
print("Top 5 Channels by Subscribers:")
print(top_subscribers[["Youtuber", "subscribers"]])
```

```python
# Calculate the correlation between subscribers and video views
correlation = data["subscribers"].corr(data["video views"])
print("Correlation between Subscribers and Video Views:", correlation)
```

```python
# Group channels by category and calculate average subscribers and video views
category_stats = data.groupby("category").agg({"subscribers": "mean", "video views": "mean"})
print("Average Subscribers and Video Views by Category:")
print(category_stats)
```

```python
# Find the country with the highest average subscribers
highest_avg_subscribers_country = data.groupby("Country")["subscribers"].mean().idxmax()
print("Country with Highest Average Subscribers:", highest_avg_subscribers_country)
```

```python
# Calculate the total earnings for the dataset
total_earnings = data[["lowest_yearly_earnings", "highest_yearly_earnings"]].sum().sum()
print("Total Earnings:", total_earnings)
```

```python
# Calculate the percentage of channels by type (individual vs. brand)
channel_type_distribution = data["channel_type"].value_counts(normalize=True) * 100
print("Channel Type Distribution:")
print(channel_type_distribution)
```

```python
# Calculate the average unemployment rate for channels in each country
country_avg_unemployment = data.groupby("Country")["Unemployment rate"].mean()
print("Average Unemployment Rate by Country:")
print(country_avg_unemployment)
```

```python
# Calculate the average subscribers for channels created in each year
yearly_avg_subscribers = data.groupby("created_year")["subscribers"].mean()
print("Yearly Average Subscribers:")
print(yearly_avg_subscribers)
```

```python
# Calculate the total subscribers gained in the last 30 days
total_subscribers_last_30_days = data["subscribers_for_last_30_days"].sum()
print("Total Subscribers Gained in the Last 30 Days:", total_subscribers_last_30_days)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("Global_YouTube_Statistics.csv", encoding="latin-1")
```

```python
# YouTube Analytics
# Calculate correlations between various metrics (subscribers, video views, etc.)
correlation_matrix = data.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python

# Content Strategy
# Calculate average upload frequency per category
category_upload_frequency = data.groupby("category")["uploads"].mean()
print("Average Upload Frequency by Category:")
print(category_upload_frequency)

```

```python
# Regional Influencers
# Find the top influencers from each country based on subscribers
top_influencers_by_country = data.groupby("Country").apply(lambda group: group.nlargest(1, "subscribers"))
print("Top Influencers by Country:")
print(top_influencers_by_country[["Youtuber", "subscribers", "Country"]])

```

```python
# Earnings Analysis
# Calculate correlation between subscribers and earnings
earnings_correlation = data["subscribers"].corr(data["highest_yearly_earnings"])
print("Correlation between Subscribers and Earnings:", earnings_correlation)
```

```python
# Geospatial Visualization
# Visualize the distribution of channels on a world map
plt.figure(figsize=(12, 6))
sns.scatterplot(x="Longitude", y="Latitude", data=data, hue="subscribers", size="subscribers", sizes=(20, 200))
plt.title("Geospatial Distribution of YouTube Channels")
plt.show()

```

```python
# Trending Topics
# Analyze how categories' popularity changes over time
category_popularity_over_time = data.groupby(["created_year", "category"])["subscribers"].sum().unstack().fillna(0)
plt.figure(figsize=(10, 6))
sns.heatmap(category_popularity_over_time, cmap="YlGnBu")
plt.title("Category Popularity Over Time")
plt.xlabel("Category")
plt.ylabel("Year")
plt.show()
```

```python

# Trending Topics

trending_categories = data.groupby(["created_year", "category"]).size().unstack().fillna(0)
sns.set()
plt.figure(figsize=(12, 8))
sns.heatmap(trending_categories, cmap="YlGnBu")
plt.title("Trending Categories Over Years")
plt.xlabel("Category")
plt.ylabel("Year")
plt.show()
```
